```python
import matplotlib.pyplot as plt

# D liu mi
n_values = [500, 1000, 2000]
sync_times = [311, 614, 1819]
thread_4_times = [105, 161, 388]
thread_8_times = [58, 101, 218]

with plt.xkcd():

    # V  th
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    ax.spines[['top', 'right']].set_visible(False)
    ax.plot(n_values, sync_times, marker='o', linestyle='-', label='Synchronous', linewidth=2)
    ax.plot(n_values, thread_4_times, marker='s', linestyle='--', label='Thread(4)', linewidth=2)
    ax.plot(n_values, thread_8_times, marker='^', linestyle='-.', label='Thread(8)', linewidth=2)

    # Cu hnh  th
    plt.xlabel('Number of Tasks (n)', fontsize=12)
    plt.ylabel('Execution Time (seconds)', fontsize=12)
    plt.title('Execution Time Comparison: Synchronous vs Multithreading', fontsize=14, fontweight='bold')
    plt.xticks(n_values, fontsize=10)
    plt.yticks(fontsize=10)
    plt.legend(fontsize=10)
plt.show()
```
